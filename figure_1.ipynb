```python
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
```

```python
## S2EF energy MAE plateau plot
# source: https://opencatalystproject.org/viz/s2ef.html
model_names = ["DimeNet++", "SpinConv", "GemNet-T", "GemNet-XL", "GemNet-OC-Large", 
               "SCN", "eSCN","EquiformerV2"]
submission_dates = ["02/2021", "05/2021", "08/2021", "09/2021", "01/2022", 
                   "05/2022","03/2023", "05/2023"]
energy_maes = [0.5342, 0.3363, 0.2924, 0.2701, 0.2411, 0.2368, 0.2281, 0.2187]
```

```python
# Convert submission dates to datetime objects
dates = [datetime.strptime(date, "%m/%Y") for date in submission_dates]

# Define colors for each model
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray']

# Plotting
plt.figure(figsize=(10, 6),dpi=400)


for model, date, energy_mae, color in zip(model_names, dates, energy_maes, colors):
    plt.scatter(date, energy_mae, s=100, alpha=0.8, label=model, color=color)

# Adjust x-axis ticks to show only January and July
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(bymonth=[1, 7]))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))

# Add labels and title
plt.xlabel('Submission Date', fontsize=18)
plt.ylabel('Energy MAE (eV)',fontsize=18)

# Add legend
plt.legend(fontsize=18)
# Increase font size for ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
# Set y-axis limits
plt.ylim(0.2, 0.55)
# Customize spines to remove top and right borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
# Remove grid
plt.grid(False)
plt.tight_layout()
plt.show()
```
